```python
with open(sourcepath,'r',encoding='utf8') as f:
    l=f.readlines()
```

```python
#在每个标注后面加空格
for i in range(len(l)):
    for k in dp.tags:
        l[i] = re.sub('/'+k,'/'+k+' ',l[i])
```

```python
l[2]
```

```python
gg = list()
for ll in [s.split() for s in l]:
    for ss in ll:
        ss = ss.strip()
        i = ss.rfind('/')
        flag = ss[i+1:]
        for c in ss[:i]:
            gg.append(flag)
    gg.append('o')
```

```python
tags = ['o','b', 'jx', 'jc', 's', 'x', 'yc',  'zs']
comments = ['o', '疾病名称', '畸形', '异常检查', '受伤',  '细菌、病毒', '异常体征',  '自诉症状']
dict1 = dict(zip(tags, comments))
```

```python
tags2 = ['t', 'yw', 'zl', 'm', 'qt','w']
comments2 = ['查体','药物', '治疗过程', '医疗设备', '其他治疗', '部位']
dict2 = dict(zip(tags2, comments2))
```

```python
#修饰
dict3 = {'2':'非患者本人',
'3':'有条件的',
'4':'可能的',
'5':'待证实的',
'6':'偶有的',
'7':'既往的',
'1':'否认的'}
```

```python
t = 0 #记录entity数
e = 0 #记录events数
a = 0 #记录修饰的数量
pre = 'o'
preindex = 0
s = ''
g1 = list()
for i in range(len(gg)):
    if gg[i] != pre and pre != 'o':
        t += 1
        if pre[0] in list('1234567'):
            a += 1
            s = 'A{}\t{} T{}'.format(a,dict3[pre[0]],t)
            pre = pre[1:]
        if pre in dict1:
            g1.append('T{}\t{} {} {}\t{}'.format(t,dict1[pre],preindex,i,text[preindex:i]))
        elif pre in dict2:
            e += 1
            g1.append('T{}\t{} {} {}\t{}'.format(t,dict2[pre],preindex,i,text[preindex:i]))
            g1.append('E{}\t{}:T{} '.format(e,dict2[pre],t))
        if s:
            g1.append(s)
            s = ''
        pre = 'o'
    if gg[i] != 'o' and gg[i] != pre:
        pre = gg[i]
        preindex = i
```

```python
with open('a.txt','w',encoding='utf8') as f:
    f.write('\n'.join(g1))
```
