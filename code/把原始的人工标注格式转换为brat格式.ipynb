### 代码重构

### 要用到的字典

```python
#标注器左侧的tag  
tags = ['o','b', 'jx', 'jc', 's', 'x', 'yc',  'zs']
comments = ['o', '疾病名称', '畸形', '异常检查', '受伤',  '细菌、病毒', '异常体征',  '自诉症状']
dict1 = dict(zip(tags, comments))
#标注器右侧的tag  
tags2 = ['t', 'yw', 'zl', 'm', 'qt','w']
comments2 = ['查体','药物', '治疗过程', '医疗设备', '其他治疗', '部位']
dict2 = dict(zip(tags2, comments2))
#修饰
dict3 = {'2':'非患者本人',
'3':'有条件的',
'4':'可能的',
'5':'待证实的',
'6':'偶有的',
'7':'既往的',
'1':'否认的'}
```

```python
def f1(l):
    '''
    将人工标注文件转换为taglist。
    如有/o糖尿病史/b多年  转换为 
    ['o', 'b', 'b', 'b', 'b', 'o', 'o', 'o',
    
    l: 文章的行列表
    
    return: 
    '''
    #在每个标注后面加空格
    for i in range(len(l)):
        for k in dp.tags:
            l[i] = re.sub('/'+k,'/'+k+' ',l[i])
    gg = list()
    for ll in [s.split() for s in l]:
        for ss in ll:
            ss = ss.strip()
            i = ss.rfind('/')
            s1 = ''
            if i == -1:#主要针对的是标注时，行结尾忘记标\o的情况
                flag = 'o'
                s1 = ss
            else:
                flag = ss[i+1:]
                flag1 = flag
                if flag[0] in list('1234567'):
                    flag1 = flag[1:]            
                if flag1 not in dict1 and flag1 not in dict2:
                    flag = 'o'
                    s1 = ss
                else:
                    s1 = ss[:i]
            for c in s1:
                gg.append(flag)
        gg.append('o')
    return gg
def f2(gg,text):
    '''
    把gg转换成brat标注的格式
    '''
    t = 0 #记录entity数
    e = 0 #记录events数
    a = 0 #记录修饰的数量
    pre = 'o'
    preindex = 0
    s = ''
    g1 = list()
    for i in range(len(gg)):
        if gg[i] != pre and pre != 'o':
            t += 1
            tt = ''
            if pre[0] in list('1234567'):
                a += 1
                tt = pre[0]
                pre = pre[1:]
            if pre in dict1:
                g1.append('T{}\t{} {} {}\t{}'.format(t,dict1[pre],preindex,i,text[preindex:i]))
                if tt:
                    s = 'A{}\t{} T{}'.format(a,dict3[tt],t)
                    tt = ''
            elif pre in dict2:
                e += 1
                g1.append('T{}\t{} {} {}\t{}'.format(t,dict2[pre],preindex,i,text[preindex:i]))
                g1.append('E{}\t{}:T{} '.format(e,dict2[pre],t))
                if tt:
                    s = 'A{}\t{} E{}'.format(a,dict3[tt],e)
                    tt = ''
            if s:
                g1.append(s)
                s = ''
            pre = 'o'
        if gg[i] != 'o' and gg[i] != pre:
            pre = gg[i]
            preindex = i
    return g1
def f3(g1, destfile):
    '''
    把g1输出成brat的ann文件
    
    destfile: 输出目标文件路径
        如 a/b.ann
    '''
    with open(destfile,'w',encoding='utf8') as f:
        f.write('\n'.join(g1))    
def ff(rawfile, manfile, destfile):
    '''
    rawfile: 原始未标注的文件
    manfile: 人工标注的文件
    destfile: 输出的ann文件
    '''
    #读取无标注文件
    with open(rawfile,'r',encoding='utf8') as f:
        text=f.read()
    #读取人工标注文件
    with open(manfile,'r',encoding='utf8') as f:
        l=f.readlines()
    #将人工标注文件转换为taglist
    gg = f1(l)
    #把gg转换成brat标注的格式
    g1 = f2(gg,text)
    # 把g1输出成brat的ann文件
    f3(g1, destfile)
```
