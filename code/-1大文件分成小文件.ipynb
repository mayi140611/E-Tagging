```python
df1 = df[['ZS','XBS','TGJC']]
```

```python
df1 = df[['ZS','XBS','TGJC']].fillna("")
```

```python
import re
def catfields(x):
    for f in ['ZS', 'XBS', 'TGJC']:
        x[f] = re.sub('\\r\\n','',x[f])
    if x['ZS']:
        if re.match(r'{}'.format(x['ZS']),x['XBS']):
            if x['XBS'].endswith('。') or x['XBS'].endswith('.'):
                return '{}{}'.format(x['XBS'],x['TGJC'])
            else:
                return '{}。{}'.format(x['XBS'],x['TGJC'])
        elif not x['ZS'].endswith('。'):
            if x['XBS'].endswith('。') or x['XBS'].endswith('.'):
                return '{}。{}{}'.format(x['ZS'],x['XBS'],x['TGJC'])
            else:
                return '{}。{}。{}'.format(x['ZS'],x['XBS'],x['TGJC'])
        else:
            if x['XBS'].endswith('。') or x['XBS'].endswith('.'):
                return '{}{}{}'.format(x['ZS'],x['XBS'],x['TGJC'])
            else:
                return '{}{}。{}'.format(x['ZS'],x['XBS'],x['TGJC'])
    if x['XBS'].endswith('。') or x['XBS'].endswith('.'):
        return '{}{}'.format(x['XBS'],x['TGJC'])
    else:
        return '{}。{}'.format(x['XBS'],x['TGJC'])

df1['ZS_XBS'] = df1.apply(catfields, axis=1)

df1['ZS_XBS'].to_csv('d:\\Desktop\\cndata\\ru1_tb_cis_mzdzbl.csv',index=False,header=False)
```

```python
def catfields1(x):
    for f in ['ZS', 'XBS']:
        x[f] = re.sub('\\r\\n','',x[f])
    if x['ZS']:
        if not re.match(r'{}'.format(x['ZS']),x['XBS']):
            if not x['ZS'].endswith('。'):
                return '{}。{}'.format(x['ZS'],x['XBS'])
            else:
                return '{}{}'.format(x['ZS'],x['XBS'])
    return x['XBS']
    
df1['ZS_XBS'] = df1.apply(catfields1, axis=1)

df1['ZS_XBS'].to_csv('d:\\Desktop\\cndata\\ru1_tb_cis_mzdzbl.csv',index=False,header=False)
```

```python
import os
def splitText1(dir,file):
	"""
	把大文件转化为20行的小文件
	"""
	df = pd.read_csv('{}{}.csv'.format(dir,file),encoding='utf8',header=None)
	if not os.path.exists('{}{}'.format(dir,file)): 
		os.mkdir('{}{}'.format(dir,file))
	count = 0
	while count*20<df.shape[0]:
		df.iloc[count*20:(count*20+20),:].to_csv('{0}{1}\\{1}_{2}.csv'.format(dir,file,count),index=False,header=False,sep='\n')
		count += 1
```
