```python
l = len(tags)
for t in range(1,l):
    for i in range(1,8):
        tags.append('{}{}'.format(i,tags[t]))
        if i == 1:
            s.append('{}(-)'.format(s[t]))
        elif i == 2:
            s.append('{}(家族成员)'.format(s[t]))
        elif i == 3:
            s.append('{}(有条件的)'.format(s[t]))
        elif i == 4:
            s.append('{}(可能的)'.format(s[t]))
        elif i == 5:
            s.append('{}(待证实的)'.format(s[t]))
        elif i == 6:
            s.append('{}(偶有的)'.format(s[t]))
        elif i == 7:
            s.append('{}(既往史)'.format(s[t]))
print(tags)
s
```
